```python
# Biblioteca seaborn cria graficos para visualização 
# Biblioteca matplotlib.pyplot para a comunicação com o searbon
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregando o DataSet
iris = sns.load_dataset('iris')
```

```python
# Mostrando as 5 primeiras linhas do DataSet 
iris.head()
```

```python
# Mostrando as especies e a quantidade de cada especie
iris['species'].value_counts()
```

```python
iris.head()
```

```python
iris.describe()
```

```python
# Mostra um grafico chamado de histograma que mostra a frequencia dos valores continuos
col = 'sepal_length'
sns.histplot(data=iris, x='sepal_length', kde = True).set_title(f'Distribuição da variavel {col}')
```

```python
for col in iris.drop(columns='species'):
    sns.histplot(data=iris, x=col, kde=True).set_title(f'Distribuição da variavel {col}')
    plt.show()
```

```python
# Foi passado um parametro pedindo que os graficos ficassem com cor para cada especie
for col in iris.drop(columns='species'):
    sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title(f'Distribuição da variavel {col}')
    plt.show()
```

```python
# Barplot é um tipo de visualização de dados que ultiliza barras retangulares
# Ci é um intervalo de de confiança ou intervalo numerico que estima , com um determinaso nível de confiança
for col in iris.drop(columns='species'):
    sns.barplot(data=iris, x=iris['species'], hue=iris['species'], y = col, ci = 90)
    plt.show()
```
